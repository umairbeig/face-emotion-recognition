```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import imread
from matplotlib.pyplot import imshow
import cv2
import os
```

```python
classes=["angry","disgust","fear","happy","neutral","sad","surprise"]
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
new_model=tf.keras.models.load_model("drive/MyDrive/major project/modelumair1111.h5")
```

```python
hpp=cv2.imread("drive/MyDrive/major project/faces/5.jpg")
imshow(hpp)
```

```python
face_casc=cv2.CascadeClassifier("drive/MyDrive/major project/haarcascade_frontalface_default.xml")
```

```python
gray=cv2.cvtColor(hpp,cv2.COLOR_BGR2GRAY)
```

```python
imshow(gray)
```

```python
gray.shape
```

```python
faces = face_casc.detectMultiScale(gray, 1.3, 5)

for (x,y,w,h) in faces:
    cv2.rectangle(hpp,(x,y),(x+w,y+h),(0,0,255),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = hpp[y:y+h, x:x+w]
    faces=face_casc.detectMultiScale(roi_gray)
    if len(faces)==0:
      print("no face detected")
    else:
      for (ex,ey,ew,eh)in faces:
        face_roi=roi_color[ey:ey+eh,ex:ex+ew]
```

```python
imshow(hpp)
```

```python
x=cv2.cvtColor(face_roi,cv2.COLOR_BGR2RGB)
x.shape
```

```python
import numpy as np
final_img=cv2.resize(x,(224,224))
final_img=np.expand_dims(final_img,axis=0)
# final_img=final_img/255.0
```

```python
final_img.shape
```

```python
pred=new_model.predict(final_img)
pred
```

```python
classes[np.argmax(pred)]
```
